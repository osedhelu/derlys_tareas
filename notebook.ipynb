# 1. Elabore el algoritmo y programa python que realice lo siguiente: Lea un vector de N números enteros e intercambie sus elementos en el mismo vector (no puede usar un vector auxiliar), de la siguiente forma: el primero con el último, el segundo con el penúltimo, el tercero con el antepenúltimo y así sucesivamente.
**Nota: No pueden usar la función reverse, ni el slicing para revertir la cadena.**

Ejemplo:

NUME - INICIAL
# 
 | Pos 1 | Pos 2 | Pos 3 | Pos 4 | Pos 5 | Pos 6 | Pos 7 | Pos 8 | Pos 9 |
 |-------|-------|-------|-------|-------|-------|-------|-------|-------|
 |   10  |   15  |   80  |   30  |   90  |   20  |   50  |   40  |   70  |


 
NUME - FINAL

  | Pos 1 | Pos 2 | Pos 3 | Pos 4 | Pos 5 | Pos 6 | Pos 7 | Pos 8 | Pos 9 |
  |-------|-------|-------|-------|-------|-------|-------|-------|-------|
  |   70  |   40  |   50  |   20  |   90  |   30  |   80  |   15  |   10  |


Debe imprimir el vector final

### Te explico cómo funciona el código paso a paso:
1. Definición del vector inicial:
```python vector = [10, 15, 80, 30, 90, 20, 50, 40, 70]
n = len(vector)
```
2. Mostrar el vector inicial:
``` python
print("\nVector inicial:")
for i in range(n):
    print(f"Posición {i+1}: {vector[i]}")
```
- Imprime un título "Vector inicial"
- Usa un bucle `for` para recorrer cada posición del vector
- `range(n)` genera números del 0 al 8 (porque n=9)
- `i+1` se usa para mostrar las posiciones desde 1 en lugar de 0
- `vector[i]` accede al elemento en la posición i
3. Cálculo de la mitad:
``` python
mitad = n // 2
```
- `n // 2` hace una división entera de 9 entre 2
- El resultado es 4 (porque 9/2 = 4.5, y la división entera toma solo la parte entera)
- Esto nos dice que necesitamos hacer 4 intercambios

4. Proceso de inversión:
``` python
for i in range(mitad):
    pos_opuesta = n - 1 - i
    
    temp = vector[i]
    vector[i] = vector[pos_opuesta]
    vector[pos_opuesta] = temp
```

- `for i in range(mitad)`: recorre desde 0 hasta 3 (4 iteraciones)
- `pos_opuesta = n - 1 - i`: calcula la posición opuesta
    - Para i=0: pos_opuesta = 9-1-0 = 8
    - Para i=1: pos_opuesta = 9-1-1 = 7
    - Para i=2: pos_opuesta = 9-1-2 = 6
    - Para i=3: pos_opuesta = 9-1-3 = 5
- El intercambio se hace usando una variable temporal `temp`:
    1. Guardamos el valor de `vector[i]` en `temp`
    2. Copiamos el valor de `vector[pos_opuesta]` a `vector[i]`
    3. Copiamos el valor de `temp` a `vector[pos_opuesta]`

5. Mostrar el vector final:
``` python
print("\nVector final:")
for i in range(n):
    print(f"Posición {i+1}: {vector[i]}")
```
- Similar al paso 2, pero ahora muestra el vector ya invertido


Ejemplo de cómo funciona el intercambio:

``` text 
Vector inicial: [10, 15, 80, 30, 90, 20, 50, 40, 70]

Primer intercambio (i=0):
- Intercambia posición 0 (10) con posición 8 (70)
Resultado: [70, 15, 80, 30, 90, 20, 50, 40, 10]

Segundo intercambio (i=1):
- Intercambia posición 1 (15) con posición 7 (40)
Resultado: [70, 40, 80, 30, 90, 20, 50, 15, 10]

Tercer intercambio (i=2):
- Intercambia posición 2 (80) con posición 6 (50)
Resultado: [70, 40, 50, 30, 90, 20, 80, 15, 10]

Cuarto intercambio (i=3):
- Intercambia posición 3 (30) con posición 5 (20)
Resultado: [70, 40, 50, 20, 90, 30, 80, 15, 10]

```

````python
# Programa para invertir un vector sin usar funciones predefinidas ni vector auxiliar

# Vector predefinido
vector = [10, 15, 80, 30, 90, 20, 50, 40, 70]
n = len(vector)

# Imprimir el vector inicial
print("\nVector inicial:")
for i in range(n):
    print(f"Posición {i+1}: {vector[i]}")

# Calcular la mitad del vector
mitad = n // 2

# Intercambiar los elementos
for i in range(mitad):
    # Calcular la posición opuesta
    pos_opuesta = n - 1 - i
    
    # Intercambiar los elementos
    temp = vector[i]
    vector[i] = vector[pos_opuesta]
    vector[pos_opuesta] = temp

# Imprimir el vector final
print("\nVector final:")
for i in range(n):
    print(f"Posición {i+1}: {vector[i]}")
````

- C (i) = A(i) + B(i) – 2 -> Si i es par 
- C (i) = A(i) - B(i) + 7 -> Si i es impar

Imprima las tres listas.

### Pasos de ejecucion a realizar 
El ejercicio pide:
1. Tener dos listas A y B con M elementos
2. Crear una tercera lista C donde:
    - Si la posición es par: C[i] = A[i] + B[i] - 2
    - Si la posición es impar: C[i] = A[i] - B[i] + 7
3. Imprimir las tres listas
Voy a crear un nuevo archivo con la solución.


### Te explico cómo funciona el código paso a paso:
1. Listas iniciales:
``` python
A = [10, 15, 20, 25, 30]
B = [5, 10, 15, 20, 25]
```
- Creamos dos listas de ejemplo con 5 elementos cada una

2. Tamaño de las listas:
```python
M = len(A)
```
- Obtenemos el tamaño de las listas (en este caso 5)

3. Crear lista C:
```python
C = [0] * M
```

- Creamos una lista C del mismo tamaño que A y B
- Inicialmente todos sus elementos son 0

4. Cálculo de C:
``` python
for i in range(M):
    if i % 2 == 0:  # Si es par
        C[i] = A[i] + B[i] - 2
    else:  # Si es impar
        C[i] = A[i] - B[i] + 7
```

- Recorremos cada posición de 0 a 4
- Si la posición es par (0, 2, 4): C[i] = A[i] + B[i] - 2
- Si la posición es impar (1, 3): C[i] = A[i] - B[i] + 7


5. Imprimir resultados:
``` python
print("\nLista A:")
for i in range(M):
    print(f"Posición {i+1}: {A[i]}")
```
Cuando ejecutes el programa, verás:

```text
Posición 2: 10
Posición 3: 15
Posición 4: 20
Posición 5: 25
```

````python
# Programa para generar una lista C a partir de dos listas A y B

# Listas predefinidas de ejemplo
A = [10, 15, 20, 25, 30]
B = [5, 10, 15, 20, 25]

# Obtener el tamaño de las listas
M = len(A)

# Crear lista C vacía
C = [0] * M

# Calcular los elementos de C
for i in range(M):
    # Verificar si la posición es par o impar
    if i % 2 == 0:  # Si es par
        C[i] = A[i] + B[i] - 2
    else:  # Si es impar
        C[i] = A[i] - B[i] + 7

# Imprimir las tres listas
print("\nLista A:")
for i in range(M):
    print(f"Posición {i+1}: {A[i]}")

print("\nLista B:")
for i in range(M):
    print(f"Posición {i+1}: {B[i]}")

print("\nLista C:")
for i in range(M):
    print(f"Posición {i+1}: {C[i]}")
````

# 4.Elabore el algoritmo y el programa Python que realice lo siguiente: Lea una lista de N números y devuelva en otra lista la suma acumulada, es decir una nueva lista donde el primer elemento es el mismo, el segundo elemento es la suma del primero con el segundo, el tercer elemento es la suma del resultado anterior con el siguiente elemento y así sucesivamente. Por ejemplo, la suma acumulada de [1,2,3] es [1, 3, 6]

### Te explico cómo funciona el código paso a paso:
1. Lista inicial:
``` python 
lista = [1, 2, 3, 4, 5]
```
- Creamos una lista de ejemplo con 5 números
2. Crear lista para la suma acumulada:
``` python
suma_acumulada = [0] * len(lista)
```
- Creamos una lista del mismo tamaño que la original
- Inicialmente todos sus elementos son 0
3. Cálculo de la suma acumulada:
``` python
suma_actual = 0
for i in range(len(lista)):
    suma_actual = suma_actual + lista[i]
    suma_acumulada[i] = suma_actual
```


- `suma_actual` guarda la suma hasta el momento
- Para cada posición:
    - Sumamos el número actual a `suma_actual`
    - Guardamos el resultado en `suma_acumulada[i]`

4. Imprimir resultados:
```python
print("\nLista original:")
for i in range(len(lista)):
    print(f"Posición {i+1}: {lista[i]}")

print("\nLista con suma acumulada:")
for i in range(len(suma_acumulada)):
    print(f"Posición {i+1}: {suma_acumulada[i]}")
```

```text
Posición 2: 3   (1 + 2)
Posición 3: 6   (3 + 3)
Posición 4: 10  (6 + 4)
Posición 5: 15  (10 + 5)
```


````python
# Programa para calcular la suma acumulada de una lista

# Lista predefinida de ejemplo
lista = [1, 2, 3, 4, 5]

# Crear lista para la suma acumulada
suma_acumulada = [0] * len(lista)

# Calcular la suma acumulada
suma_actual = 0
for i in range(len(lista)):
    suma_actual = suma_actual + lista[i]
    suma_acumulada[i] = suma_actual

# Imprimir resultados
print("\nLista original:")
for i in range(len(lista)):
    print(f"Posición {i+1}: {lista[i]}")

print("\nLista con suma acumulada:")
for i in range(len(suma_acumulada)):
    print(f"Posición {i+1}: {suma_acumulada[i]}")
````

Elabore un programa Python que genere e imprima un tercer vector llamado ganados con los códigos de los estudiantes que ganaron ambas asignaturas.

1. Vectores iniciales:
```python
matematicas = [1001, 1002, 1003, 1004, 1005]
fisica = [1002, 1004, 1006, 1007, 1008]
```


- Creamos dos listas con códigos de estudiantes
- En este ejemplo:
    - Matemáticas: 5 estudiantes (1001-1005)
    - Física: 5 estudiantes (1002-1008)
2. Lista para resultados:

``` python
ganados = []
```
- Creamos una lista vacía para guardar los códigos de estudiantes que ganaron ambas asignaturas

3. Búsqueda de coincidencias:


``` python
for codigo in matematicas:
```

````python
# Programa para encontrar estudiantes que ganaron ambas asignaturas

# Vectores predefinidos de ejemplo
matematicas = [1001, 1002, 1003, 1004, 1005]
fisica = [1002, 1004, 1006, 1007, 1008]

# Crear lista para los estudiantes que ganaron ambas asignaturas
ganados = []

# Encontrar estudiantes que ganaron ambas asignaturas
for codigo in matematicas:
    # Verificar si el código está en física
    for codigo_fisica in fisica:
        if codigo == codigo_fisica:
            ganados.append(codigo)
            break  # Salir del bucle interno si encontramos coincidencia

# Imprimir resultados
print("\nEstudiantes que ganaron Matemáticas:")
for i in range(len(matematicas)):
    print(f"Posición {i+1}: {matematicas[i]}")

print("\nEstudiantes que ganaron Física:")
for i in range(len(fisica)):
    print(f"Posición {i+1}: {fisica[i]}")

print("\nEstudiantes que ganaron ambas asignaturas:")
if len(ganados) > 0:
    for i in range(len(ganados)):
        print(f"Posición {i+1}: {ganados[i]}")
else:
    print("No hay estudiantes que hayan ganado ambas asignaturas")
````

# 6. En una tienda que vende electrodomésticos trabajan N vendedores. Se guarda en un vector el valor total de las ventas realizadas por cada vendedor en un mes. Se paga una comisión a los vendedores dependiendo del valor de las ventas de la siguiente forma:
- Si las ventas son menores a 500.000, la comisión es del 5%
- Si las ventas están entre 500.000 y 800.000 (inclusive), la comisión es del 8%
- Si las ventas son superiores a 800.000, la comisión es del 10%
Desarrollar un algoritmo y programa en Python para:
1. Imprimir el valor de la comisión ganada por cada vendedor
2. Imprimir el promedio de comisión de la tienda
3. Imprimir el número de vendedores que obtuvieron comisión por encima del promedio de la tienda
4. Imprimir el valor total de ventas de los vendedores cuya comisión es mayor al promedio de la tienda

### Te explico cómo funciona el código paso a paso:
1. Lista de ventas:
``` python
ventas = [450000, 600000, 900000, 300000, 750000]
```

- Creamos una lista con las ventas de 5 vendedores como ejemplo
2. Cálculo de comisiones:
``` python
comisiones = []
for venta in ventas:
    if venta < 500000:
        comision = venta * 0.05  # 5%
    elif venta <= 800000:
        comision = venta * 0.08  # 8%
    else:
        comision = venta * 0.10  # 10%
    comisiones.append(comision)
```
- Para cada vendedor, calculamos su comisión según el rango de ventas
- Guardamos las comisiones en una nueva lista

3. Cálculo del promedio:
``` python
promedio_comision = 0
for comision in comisiones:
    promedio_comision = promedio_comision + comision
promedio_comision = promedio_comision / len(comisiones)

```
- Sumamos todas las comisiones
- Dividimos entre el número de vendedores para obtener el promedio

4. Vendedores sobre el promedio:
``` python
vendedores_sobre_promedio = 0
ventas_sobre_promedio = 0

for i in range(len(comisiones)):
    if comisiones[i] > promedio_comision:
        vendedores_sobre_promedio = vendedores_sobre_promedio + 1
        ventas_sobre_promedio = ventas_sobre_promedio + ventas[i]
```
- Contamos cuántos vendedores tienen comisión mayor al promedio
- Sumamos las ventas de estos vendedores
5. Imprimir resultados:
``` python
print("\nComisiones por vendedor:")
for i in range(len(comisiones)):
    print(f"Vendedor {i+1}: ${comisiones[i]:,.2f}")
```

- Mostramos las comisiones de cada vendedor
- Mostramos el promedio de comisiones
- Mostramos el número de vendedores sobre el promedio
- Mostramos el total de ventas de los vendedores sobre el promedio
Cuando ejecutes el programa, verás algo como:

``` text
Comisiones por vendedor:
Vendedor 1: $22,500.00
Vendedor 2: $48,000.00
Vendedor 3: $90,000.00
Vendedor 4: $15,000.00
Vendedor 5: $60,000.00

Promedio de comisión de la tienda: $47,100.00
Número de vendedores con comisión mayor al promedio: 2
Total de ventas de vendedores con comisión mayor al promedio: $1,650,000.00
```

````python
# Programa para calcular comisiones de vendedores

# Lista predefinida de ventas de ejemplo (en pesos)
ventas = [450000, 600000, 900000, 300000, 750000]

# Calcular comisiones
comisiones = []
for venta in ventas:
    if venta < 500000:
        comision = venta * 0.05  # 5%
    elif venta <= 800000:
        comision = venta * 0.08  # 8%
    else:
        comision = venta * 0.10  # 10%
    comisiones.append(comision)

# Calcular promedio de comisiones
promedio_comision = 0
for comision in comisiones:
    promedio_comision = promedio_comision + comision
promedio_comision = promedio_comision / len(comisiones)

# Contar vendedores con comisión mayor al promedio
vendedores_sobre_promedio = 0
ventas_sobre_promedio = 0

for i in range(len(comisiones)):
    if comisiones[i] > promedio_comision:
        vendedores_sobre_promedio = vendedores_sobre_promedio + 1
        ventas_sobre_promedio = ventas_sobre_promedio + ventas[i]

# Imprimir resultados
print("\nComisiones por vendedor:")
for i in range(len(comisiones)):
    print(f"Vendedor {i+1}: ${comisiones[i]:,.2f}")

print(f"\nPromedio de comisión de la tienda: ${promedio_comision:,.2f}")
print(f"Número de vendedores con comisión mayor al promedio: {vendedores_sobre_promedio}")
print(f"Total de ventas de vendedores con comisión mayor al promedio: ${ventas_sobre_promedio:,.2f}")
````

````python

N = int(input("Ingrese el número de términos de Fibonacci a generar: "))

fibonacci = [0] * N

if N >= 1:
    fibonacci[0] = 0
if N >= 2:
    fibonacci[1] = 1

for i in range(2, N):
    fibonacci[i] = fibonacci[i-1] + fibonacci[i-2]

suma_pares = 0
for numero in fibonacci:
    if numero % 2 == 0:
        suma_pares = suma_pares + numero

print("\nSerie de Fibonacci generada:")
for i in range(N):
    print(f" {fibonacci[i]}")

print(f"\nSuma de los elementos pares: {suma_pares}") 
````
